**DATESET 2 : WPP**

```python
WPP = pd.read_csv("WPP2024_ByAge.csv", low_memory=False) # To avoid data type mismatches in large datasets.
```

```python
WPP.columns
```

```python
WPP_filtered = WPP[WPP['Location'].isin(['United Kingdom', 'United Arab Emirates', 'Malaysia'])] # Filter rows where 'Location' is UK, UAE, or Malaysia
# I am ignoring the 'United Kingdom (and dependencies)' and only considering 'United Kingdom' for clarity
```

```python
WPP_filtered
```

```python
for col in WPP_filtered.columns: # Making sure if each column is relevant or not
    print(f"Unique values in '{col}':")
    print(WPP_filtered[col].unique(), "\n")
```

```python
columns_to_keep = ['Location', 'Time', 'MidPeriod', 'AgeGrpStart', 'PopMale', 'PopFemale', 'PopTotal']  # Keeping these columns and removing the irrelevant ones
WPP_filtered = WPP_filtered[columns_to_keep]
WPP_filtered # Population is in thousands according to the source of data
```

```python
WPP_filtered2 = WPP_filtered #Duplicating the dataframe so we dont mix them later on
WPP_filtered2 = WPP_filtered2.groupby(['Location','Time','AgeGrpStart']).sum()
WPP_filtered2 = WPP_filtered2[['PopTotal']]
WPP_filtered2 = WPP_filtered2.reset_index()
```

*I will seperate each country and apply their own age restrictions and join the data back for the working population

  **Malaysia Working Age Population:**
- Definition: 15-64 years
- Source: Department of Statistics Malaysia (DOSM), World Bank, ILO

  **United Kingdom (UK) Working Age Population:**
- Definition: 16-64 years (UK uses 16 instead of 15)
- Source: Office for National Statistics (ONS), UK Government, World Bank

  **United Arab Emirates (UAE) Working Age Population:**
- Definition: 15-64 years
- Source: Federal Competitiveness and Statistics Centre (FCSC), UAE Ministry of Human Resources and Emiratisation (MOHRE), World Bank*

```python
malaysia_data = WPP_filtered2[WPP_filtered2["Location"] == "Malaysia"]
malaysia_data
```

```python
malaysia_data = malaysia_data[(malaysia_data['AgeGrpStart'] >= 15) & (malaysia_data['AgeGrpStart'] <= 64)]
malaysia_data
```

```python
print(WPP_filtered2.head())
```

```python
UAE_data = WPP_filtered2[WPP_filtered2["Location"] == "United Arab Emirates"]
UAE_data = UAE_data[(UAE_data['AgeGrpStart'] >= 15) & (UAE_data['AgeGrpStart'] <= 64)]
UAE_data
```

```python
UK_data = WPP_filtered2[WPP_filtered2["Location"] == "United Kingdom"]
UK_data = UK_data[(UK_data['AgeGrpStart'] >= 16) & (UK_data['AgeGrpStart'] <= 64)]
UK_data
```

```python
WPP_filtered2 = pd.concat([malaysia_data, UK_data, UAE_data], ignore_index=True)
```

```python
WPP_filtered2
```

**"**************************************************************"**

```python
WPP_filtered2 = WPP_filtered2.reset_index()

WPP_filtered2 = WPP_filtered2[(WPP_filtered2['AgeGrpStart'] >= 15) & (WPP_filtered2['AgeGrpStart'] <= 60)]

WPP_filtered2
```

```python
WPP_grouped = WPP_filtered2.groupby(["Location", "Time"]).sum("PopTotal").reset_index()
WPP_grouped
```

```python
plt.figure(figsize=(12, 6)) # Set figure size

countries = WPP_grouped["Location"].unique() # Get unique countries

for country in countries: # Plot each country's data
    data = WPP_grouped[WPP_grouped["Location"] == country]
    plt.plot(data["Time"], data["PopTotal"], marker="o", linestyle="-", label=country)

# Labels and title
plt.xlabel("Year")
plt.ylabel("Working Age Population (15-60)")
plt.title("Working Age Population Trend (1950-2023)")
plt.legend(title="Country")
plt.grid(True)

# Show the plot
plt.show()
```

*We assumed that working population is 15-64 for the UAE and Malaysia, 16-64 for the UK, we can see that the UK is tremendously ahead compared to the UAE and Malaysia.*

```python
def plot_population_by_age(country_name, data): #Making a function so I dont have to type the same code for each of the 3 countries
    """
    Plots a bar graph of population by age group for a specific country.

    Parameters:
        country_name (str): Name of the country to filter data for.
        data (pd.DataFrame): DataFrame containing the population data.
    """
    # Filter data for the specified country
    country_data = data[data['Location'] == country_name]

    # Create the bar graph
    plt.figure(figsize=(12, 6)) # I usually go for (6,6) but it makes it easier to interpret with a longer x axis
    plt.bar(country_data['AgeGrp'], country_data['PopTotal'], color='skyblue')

    # Add labels and title
    plt.xlabel('Age Group', fontsize=12)
    plt.ylabel('Total Population (thousands)', fontsize=12)
    plt.title(f'Population of {country_name} by Age Group (2023)', fontsize=16)
    
    plt.show() # Display the plot
```

```python
plot_population_by_age('United Arab Emirates', WPP)
```

*We can see that there are 2 peaks for the UAE, one of the peak is around the age of 8 and the other is at 32. There's a little peak from 0-12, because of the immigrants moving in and having children in the UAE, we can see a downfall at around 17 due to students going abroad for higher studies. A peak at 32 because thats when most people come to work in the UAE for financial stability.*

```python
plot_population_by_age('United Kingdom', WPP)
```

*It seems that there aren't any visibile peaks for the UK, we can notice that after ages of 55 to 58, the population starts to decline. Its constant.*

```python
plot_population_by_age('Malaysia', WPP)
```

*We can see that there is a peak for Malaysia, the peak being around the age of 30 - 31, after which it declines.*

**GDP x Population**

```python
GDP2023_duplicate = GDP2023 # Checking the dataset and just duplicating it.
GDP2023_duplicate = GDP2023_duplicate.reset_index() # Resetting index
GDP2023_duplicate
```

```python
WPP_population_by_country
```

```python
WPP_population_by_country = WPP_population_by_country.rename(columns={'Location': 'Country Name'})
WPP_population_by_country
```

```python
GDP_population_merged = GDP2023.merge(WPP_population_by_country, on="Country Name") #Merging the 2 tables
GDP_population_merged
```

```python
GDP_population_merged["GDP per Capita"] = GDP_population_merged["2023"] / GDP_population_merged["PopTotal"]
GDP_population_merged # Finding GDP per Capita for average analysis.
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for GDP
ax1.bar(GDP_population_merged["Country Name"], 
        GDP_population_merged["2023"], 
        color='skyblue', 
        label="GDP (trillion USD)")

ax1.set_xlabel("Country")
ax1.set_ylabel("GDP (trillion USD)", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")
ax1.set_xticklabels(GDP_population_merged["Country Name"], rotation=45)

# Secondary y-axis for population
ax2 = ax1.twinx()
ax2.plot(GDP_population_merged["Country Name"], 
         GDP_population_merged["PopTotal"], 
         color='red', 
         marker="o", 
         linewidth=2, 
         label="Population (millions)")

ax2.set_ylabel("Population (millions)", color="red")
ax2.tick_params(axis='y', labelcolor="red")

# Titles and grid
plt.title("GDP and Population Comparison (2023)")
fig.tight_layout()
plt.grid(axis='y', linestyle="--", alpha=0.5)

plt.show()
```

*Even though the axis are not matched, we can get an idea of how different each country's GDP is in relation to their population.*

```python
# Ensure 'Country Name' is the index
GDP_population_merged.set_index('Country Name', inplace=True)

# Plot horizontal bar chart
GDP_population_merged['GDP per Capita'].plot.barh(
    figsize=(8, 6),
    color='skyblue',
    edgecolor='black'
)

# Add labels and title
plt.xlabel("GDP per Capita (million USD)", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.title("GDP per Capita by Country (2023)", fontsize=14)

plt.show()
```

*UAE's GDP per Capita is the highest being almost 2.5 million, the UK is just below 1 and Malaysia being less than half of the UK.*
